```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color, exposure

# Step 1: Load the image
image_path = '/content/Screenshot 2023-09-28 153746.png'
image = io.imread(image_path)

# Step 2: Convert to grayscale (handle RGBA if present)
if image.shape[-1] == 4:
    image = color.rgba2rgb(image)
gray_image = color.rgb2gray(image)

# Display the original grayscale image
plt.figure(figsize=(6, 6))
plt.imshow(gray_image, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')
plt.show()

# Step 3: Plot the histogram before histogram equalization
plt.figure(figsize=(10, 5))
plt.hist(gray_image.ravel(), bins=256, histtype='step', color='black')
plt.title('Histogram Before Equalization')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()

# Step 4: Perform histogram equalization
equalized_image = exposure.equalize_hist(gray_image)

# Display the equalized image
plt.figure(figsize=(6, 6))
plt.imshow(equalized_image, cmap='gray')
plt.title('Histogram Equalized Image')
plt.axis('off')
plt.show()

# Step 5: Plot the histogram after histogram equalization
plt.figure(figsize=(10, 5))
plt.hist(equalized_image.ravel(), bins=256, histtype='step', color='black')
plt.title('Histogram After Equalization')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()
```
